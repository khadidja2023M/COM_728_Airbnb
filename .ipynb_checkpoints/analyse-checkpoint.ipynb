```python
#analyse module

import tui
import pandas as pd 

#read csv file using pandas.
def load_data_pandas(file_path):
    file_path="Airbnb_UK_2022.csv"
    data=pd.read_csv(file_path)
    return data.head()

```

```python
#This function display the menu of analyse selection.
def analyse_menu():
    print(f"""select which menu whoud you like:
    
    {"[1]":<6}:The most popular amenities that Airbnb guests are looking for.\n 
    
    {"[2]":<6}:The average price of stay in each location.\n
    
    {"[3]":<6}:The average review scores rating for each location.\n
    
    {"[4]":<6}:The average price for host is superhost and host not superhost.\n 
    
    {"[5]":<6}:Exit the programme.                
    """)   
      
    menu_selection=input('your selection: ')
    return menu_selection.strip().lower() 
```

Identifying the most popular amenities that Airbnb guests are looking for

```python
#this function display amenities in a list   
def popular_amenities(data):
    tui.started('display the  most popular amenities:')
    popular_amenities=data['amenities'].mode()
    print(f'the most popular amenities are:{popular_amenities.values}')
    tui.completed()
```

Analyse the average price of stay in each location

```python
#this function display the average price of stay in each location.      
def average_price_location(data):
    tui.started('display average price by location:')
    average_stay_location=data.groupby(['host_location'])['price'].mean()
    print(f'The average stay in each location:\n{average_stay_location}')         
    tui.completed() 
```

Analyse the average of review score rating for each location 

```python
#this function display the average review scores rating for each location.      
def average_review_location(data):
    tui.started('display the average review scores rating:')
    average_review_location=data.groupby(['host_location'])['review_scores_location'].mean()
    print(f'The average review rate score for each location is:\n{average_review_location}')
    tui.completed()
```

Analyse the average price for host is superhost and host not superhost

```python
def average_price_host_is_superhost(data):
    tui.started('display average price for host is superhost:')
    average_price=data.groupby(['host_is_superhost'])['price'].mean()
    #How many not super_host (the count)
    not_superhost_count=data.loc[data.host_is_superhost ==False,['host_is_superhost']].count()
    #How many super_host (the count)
    superhost_count=data.loc[data.host_is_superhost ==True,['host_is_superhost']].count()
    print(f"the average price of host is not super host and host is super_host is:{average_price.values}") 
    print(f"the number of host is not superhost is {not_superhost_count.values}")
    print(f"the number of host is superhost is {superhost_count.values}")
    tui.completed()   
```
