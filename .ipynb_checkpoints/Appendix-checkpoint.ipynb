```python
import tui
import pandas as pd 
import random 
import analyse 
from matplotlib import pyplot as plt



def random_property_type_price(data):
    file_path = ('Airbnb_UK_2022.csv')
    data = analyse.load_data_pandas(file_path)
    tui.started('Random property_type average price')
    property_type = data['property_type']
    count_property_type = data['property_type']

    random_property_type = random.choices(property_type, k=3)
    unique_prices = set()
    for property_type in random_property_type:
        properties= data[data['property_type'] == property_type]
        average_price =properties['price'].mean()
        unique_prices.add(average_price)
        print(f'The average price for property types:{random_property_type}are: {average_price:.2f} ')
    fig = plt.figure(figsize=(7, 4))
    plt.bar(random_property_type, average_price)
    #plt.pie(average_price, labels=random_property_type)
    #plt.legend(loc="best", bbox_to_anchor=(1.2, 1))

    plt.title("Random property types with average price")
    plt.show()
    tui.completed()

random_property_type_price(data)
```
